**Support Vector Machines (SVM)  with TF-IDF Features**

**In this code:

Load the dataset and preprocess the ratings to create sentiment labels.
Split the dataset into training and testing sets.
Create a TF-IDF vectorizer to convert the review descriptions into TF-IDF features.
Fit and transform the training text data using the TF-IDF vectorizer.
Initialize an SVM classifier with a linear kernel.
Train the SVM classifier on the TF-IDF features from the training data.
Transform the test text data using the same TF-IDF vectorizer.
Predict sentiment using the trained SVM classifier.
Calculate accuracy and print a classification report that includes precision, recall, and F1-score for both positive and negative sentiment.

```python
# Import necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score



# Preprocess the ratings to create sentiment labels
df['sentiment'] = df['rating'].apply(lambda x: 'positive' if x >= 4 else 'negative')

# Split the dataset into training and testing sets
train_data, test_data = train_test_split(df, test_size=0.2, random_state=42)

# Create the TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer(max_features=5000)  # You can adjust the max_features parameter

# Fit and transform the training text data
X_train_tfidf = tfidf_vectorizer.fit_transform(train_data['review_description'])

# Initialize the SVM classifier
svm_classifier = SVC(kernel='linear')

# Train the SVM classifier on the TF-IDF features
svm_classifier.fit(X_train_tfidf, train_data['sentiment'])

# Transform the test text data using the same vectorizer
X_test_tfidf = tfidf_vectorizer.transform(test_data['review_description'])

# Predict sentiment using the trained SVM classifier
predictions = svm_classifier.predict(X_test_tfidf)

# Calculate accuracy and print classification report
accuracy = accuracy_score(test_data['sentiment'], predictions)
print(f'Accuracy: {accuracy:.2f}')
print(classification_report(test_data['sentiment'], predictions))
```

**Here's what the output means:**

Accuracy: 0.84: This means that the SVM classifier correctly predicted the sentiment of the reviews about 84% of the time on the test set.

Classification Report:

Precision: Precision tells us how many of the predicted positive reviews are actually positive, and how many of the predicted negative reviews are actually negative. For positive sentiment, the precision is 0.83, which means that 83% of the reviews predicted as positive were actually positive. For negative sentiment, the precision is 0.85, indicating that 85% of the predicted negative reviews were actually negative.

Recall: Recall, also known as sensitivity or true positive rate, tells us how many of the actual positive reviews were correctly predicted as positive by the model, and how many of the actual negative reviews were correctly predicted as negative. For positive sentiment, the recall is 0.89, meaning that the model captured 89% of the actual positive reviews. For negative sentiment, the recall is 0.77, indicating that the model captured 77% of the actual negative reviews.

F1-score: The F1-score is a balanced measure that combines both precision and recall. It's useful when there's an uneven class distribution. For positive sentiment, the F1-score is 0.86, and for negative sentiment, it's 0.81.

Support: The support indicates the number of instances of each class in the test set. There are 3332 instances of negative sentiment reviews and 4057 instances of positive sentiment reviews.

Macro Avg: The macro average calculates the average performance across both classes. The macro average F1-score is 0.83.

Weighted Avg: The weighted average calculates the average performance, but takes into account the class distribution. The weighted average F1-score is 0.84.

Overall, the results show that the SVM classifier with TF-IDF features is performing reasonably well on this sentiment classification task. The accuracy is good, and the precision, recall, and F1-score metrics indicate balanced performance between the positive and negative sentiment classes.

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

# Create a confusion matrix
conf_matrix = confusion_matrix(test_data['sentiment'], predictions, labels=['negative', 'positive'])

# Create a heatmap for the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Negative', 'Positive'], yticklabels=['Negative', 'Positive'])
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

**SENTIMENT ANALYSIS USING RANDOM FOREST CLASSIFER**

```python
# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score


# Initialize the Random Forest classifier
random_forest_classifier = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the Random Forest classifier on the TF-IDF features
random_forest_classifier.fit(X_train_tfidf, train_data['sentiment'])

# Transform the test text data using the same vectorizer
X_test_tfidf = tfidf_vectorizer.transform(test_data['review_description'])

# Predict sentiment using the trained Random Forest classifier
predictions = random_forest_classifier.predict(X_test_tfidf)

# Calculate accuracy and print classification report
accuracy = accuracy_score(test_data['sentiment'], predictions)
print(f'Accuracy: {accuracy:.2f}')
print(classification_report(test_data['sentiment'], predictions))
```

**User Review Titles Analysis**

```python
from collections import Counter
from wordcloud import WordCloud
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')

# Remove NaN values and extract user review titles
review_titles = df['review_title'].dropna()

# Tokenize and count the words in review titles
words = ' '.join(review_titles).split()

# Define stopwords to be removed
stop_words = set(stopwords.words('english'))

# Filter out stopwords and words with length <= 2
filtered_words = [word for word in words if word not in stop_words and len(word) > 2]

# Count the occurrences of each filtered word
word_counts = Counter(filtered_words)

# Plot the most common meaningful words in review titles
common_words = word_counts.most_common(20)  # Adjust the number of common words as needed

# Extract words and their counts for plotting
words = [word for word, count in common_words]
counts = [count for word, count in common_words]

# Create a bar plot of common meaningful words in review titles
plt.figure(figsize=(10, 6))
plt.bar(words, counts, color='purple')
plt.title('Most Common Meaningful Words in User Review Titles')
plt.xlabel('Word')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

**INTERPETATION It's interesting to see that the words "App" and "Twitter" are the most frequently used words in the user review titles. This provides valuable insights into the common themes or topics that users are addressing in their review titles. The prominence of the word "App" indicates that users often refer to the app itself in their titles, while the mention of "Twitter" suggests that users might be comparing the app to or associating it with the Twitter platform.
